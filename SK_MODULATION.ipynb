# SK (Shift Keying) Modulation

### Import required libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Setting inititial parameters (Mutable)

```python
# Parameters
sampling_rate = 1000  # Hz
duration = 1.0       # sec
carrier_frequency = 50  # Hz
square_signal_frequency = 5  # Hz (Adjustable frequency for the square wave bit signal)
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)  # Time vector
```

### Input Signal (Pulse/Numeric)

```python
input_signal = 0.5 * (1 + np.sign(np.sin(2 * np.pi * square_signal_frequency * t)))
```

### Carrier Signal (Analog)

```python
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)  # Time vector
carrier_signal = np.sin(2 * np.pi * 5 * t)  # Input signal 5 Hz
```

### ASK Modulation

```python
def ask_modulation(bit_signal, carrier_frequency, sampling_rate):
    carrier = np.sin(2 * np.pi * carrier_frequency * t)
    ask_signal = bit_signal * carrier
    return ask_signal
ask_signal = ask_modulation(bit_signal, carrier_frequency, sampling_rate)
```

### FSK Modulation

```python
def fsk_modulation(bit_signal, f1, f2, sampling_rate):
    fsk_signal = np.sin(2 * np.pi * (f1 * (1 - bit_signal) + f2 * bit_signal) * t)
    return fsk_signal
fsk_signal = fsk_modulation(bit_signal, 30, 70, sampling_rate)
```

### PSK Modulation

```python
def psk_modulation(bit_signal, carrier_frequency, sampling_rate):
    psk_signal = np.sin(2 * np.pi * carrier_frequency * t + np.pi * bit_signal)
    return psk_signal
psk_signal = psk_modulation(bit_signal, carrier_frequency, sampling_rate)
```

### Ploting Signals

```python
# Plotting
plt.figure(figsize=(12, 10))
# input signal
plt.subplot(5, 1, 1)
plt.plot(t, input_signal, color='black')
plt.title("Square Wave Bit Signal")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
# carrier signal
plt.subplot(5, 1, 2)
plt.plot(t, carrier_signal, color='aqua') 
plt.title("Input Signal (5 Hz Sine Wave)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
# ASK Signal
plt.subplot(5, 1, 3)
plt.plot(t, ask_signal, color='red')
plt.title("ASK Modulation (Amplitude Shift Keying)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
# FSK Signal
plt.subplot(5, 1, 4)
plt.plot(t, fsk_signal, color='blue')
plt.title("FSK Modulation (Frequency Shift Keying)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
# PSK signal
plt.subplot(5, 1, 5)
plt.plot(t, psk_signal, color='green')
plt.title("PSK Modulation (Phase Shift Keying)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")

plt.tight_layout()
plt.show()
```
